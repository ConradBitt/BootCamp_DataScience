# Produção Hospitalar

Os dados que serão utilizados neste notebook são os do DataSUS. O [DataSUS](https://datasus.saude.gov.br/) é o sistema de dados do ministério da saúde que tem o objetivo de subsediar analises em questões de vigilancia em saúde, epidemiológica e sanitária a fim de elaborar relatórios que fundamentem tomadas de decisões que envolvem saúde publica.

Um dos provedores de dados do DataSUS é o [TabNet](https://datasus.saude.gov.br/informacoes-de-saude-tabnet/) e uma das bases de dados que o TabNet fornece é a Produção Hospitalar, no caso o [SIH/SUS](https://datasus.saude.gov.br/acesso-a-informacao/producao-hospitalar-sih-sus/)

é do tipo "dtypes: object(154)", ou seja é uma string, então ao importar os dados precisamos converte-los para números...

```python
# As vezes é necessário usar outra decodificação encoding = "ISO-8859-1"
dados = pd.read_csv('./dados/A151346189_28_143_208.csv',
                    sep=';',
                    skiprows=3,
                    skipfooter=12,
                    thousands='.',
                    decimal=',',
                    engine='python')
dados.head()
```

```python
dados.info()
```

Agora os dados são do tipo `dtypes: float64(151), object(3)` então é possível efetuar a média.

```python
dados.mean()
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
dados.mean()
```

Esta é a média de gasta por mês em cada estado.

```python
print('Gasto Mês de Agosto\n')
print('Ano  Gasto(R$)')
for ano in range(2008,2021):
    print(ano, '{:.2f}'.format(dados.mean().loc[f'{ano}/Abr']))
```

A forma de acessar elementos pelo rótulo é usando o `loc[]`. Um [DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html) é um conjunto de outra estrutura de dados chamada de [Séries](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html?highlight=series#pandas.Series), como podemos verificar:

```python
type(dados['2008/Ago'])
```

Ou seja, é um objeto do tipo Series.

# Matplotlib

O Pandas não é uma biblioteca feita para visualização, apesar de ter algumas formas basicas de visualização. É recomendavel utilizar outras bibliotecas como o [Seaborn](https://seaborn.pydata.org/), [Matplotlib](https://matplotlib.org/) ou até mesmo o [Plotly](https://plotly.com/).

## Desafio 1:
Escolher um título mais descritivo que passe uma mensagem mais adequada ao gráfico de barras.

```python
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker

ax = dados.plot(x='Unidade da Federação', y ='2008/Ago', kind='bar', figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# Titulo
plt.title("Repasse do SUS aos estados em Agosto/2008 por cada Unidade Federativa", fontsize=18, pad=20)
plt.grid(True)

ax = ax
```

# Desafio 2: 

Faça a mesma analise realizada em áula, porem para o mês mais recente

O mês mais recente é 2020/Ago, podemos verificar, desta maneira:

# Comparando mês a mês

Note que se extrair a média do gasto do estado com o SUS dentre todos os estados, vamos ter o gasto mensal

```python
dados.mean()
```

Temos o gasto de 2008/Jan depois 2008/Fev, e assim por diante... Podemos extrair só o índice deste conjunto: 

```python
dados['Unidade da Federação'].values
```

```python
meses_selecionaveis = dados.mean().index.tolist()
meses_selecionaveis.insert(0, 'Unidade da Federação')
meses_selecionaveis
```

E usar como um select no dataframe e o resultado será as colunas contidas no index:

```python
dados_tratados = dados[meses_selecionaveis]
dados_tratados
```

Note que não faz muito senido "Rondônia" ser o primeiro registro, "Acre" o segundo, esses estados não tem preferencia, o que eles deveriam ser são os índices e existe uma função chamada `set_index()`:

```python
dados_tratados = dados_tratados.set_index('Unidade da Federação')
dados_tratados
```

Agora é mais simples pois podemos buscar o registro de um dado pelo nome. 

```python
dados_tratados.plot()
```

```python
dados_tratados = dados_tratados.drop('Total', axis=1)
dados_tratados
```
