Fonte: Ministério da Saúde - Sistema de Informações Hospitalares do SUS (SIH/SUS)
 Notas:
 
 Dados referentes aos últimos seis meses, sujeitos a  atualização.
 A partir do processamento de junho de 2012, houve mudança na classificação da natureza e esfera dos estabelecimentos. Com isso, temos que:
      
* Até maio de 2012 estas informações estão disponíveis como "Natureza" e "Esfera Administrativa".
* De junho de 2012 a outubro de 2015, estão disponíveis tanto como "Natureza" e "Esfera Administrativa", como "Natureza Jurídica" e "Esfera Jurídica".
* A partir de novembro de 2015, estão disponíveis como "Natureza Jurídica" e "Esfera Jurídica".

## Heterocedasticidade

**Os dados não estão normalizados por habitantes**

```python
import numpy as np 
dados_estados_selecionados = dados_tratados.T[estados_selecionados[-4:]]
dados_estados_selecionados
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('tab10')

ax = dados_estados_selecionados[dados_estados_selecionados!=0].dropna().diff()[1:].plot(figsize=(10,6))
plt.title('Variação do repasse do SUS para 4 estados de 2008 a 2020', fontsize=18, pad=20)
#ax.yaxis.set_major_formatter(ticker.FuncFormatter(milions))
legenda = ax.legend(loc='lower left', bbox_to_anchor=(0.0, -0.2), ncol=4)
                                                                                                          
ax = ax
```

# Desafio 2 aula 3 

Adicionar mostrando uma coluna mostrando a região (Norte, Nordeste, Sul, Suldeste, Centro-Oeste) de cada estado

As macrorregiões brasileiras, são: 

<img src='https://4.bp.blogspot.com/-BnljCuSmovs/UlSM8NNq88I/AAAAAAAALfk/4uDFmbpdeuo/s1600/mapa-do-brasil-regioes.png' width=60%>

```python
#  Slice dos estados
norte = dados_tratados.index[0:7]
nordeste = dados_tratados.index[7:16]
sudeste = dados_tratados.index[15:19]
sul = dados_tratados.index[20:23]
centro_oeste = dados_tratados.index[23:]

estados = {'Norte': norte,'Nordeste': nordeste,'Sudeste': sudeste,'Sul': sul,'Centro Oeste': centro_oeste}
```

```python
for nome, estado in estados.items():
    ax = ordenado_por_total.loc[estado].T.plot(figsize=(10,6), kind='line', cmap='brg')
    plt.title(f'Repasse do SUS para a macrorregião {nome}', fontsize=18, pad=20)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(milions))
    ax.tick_params(axis='y', which='major', labelsize=14)
    
```

Um ponto curioso é que apesar da sazonalidade todos os estados seguem uma tenencia, apenas o estado do Paraná aparenta quebrar esta linha de tendencia a partir de Julho de 2016.

Extraindo a sazonalidade, com uma média móvel de 12 meses podemos verificar esta alteração na tendência.

```python
for nome, estado in estados.items():
    ax = ordenado_por_total.loc[estado].T.rolling(12).mean().plot(figsize=(10,6), kind='line', cmap='brg', )
    plt.title(f'Repasse do SUS para a macrorregião {nome}', fontsize=18, pad=20)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(milions))
    ax.tick_params(axis='y', which='major', labelsize=14)
    
```
